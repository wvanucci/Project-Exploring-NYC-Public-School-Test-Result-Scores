Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
schools_no_null = schools.fillna(0)
best_math_schools = schools_no_null[schools_no_null['average_math'] >=(800*80/100)][['school_name','average_math']].sort_values('average_math', ascending = False)
best_math_schools
# Add as many cells as you like...
```

```python
schools_no_null['total_SAT'] = (schools_no_null['average_math'] + schools_no_null['average_reading']+
                              schools_no_null['average_writing'])
top_10_schools = schools_no_null.loc[:,['school_name','total_SAT']].sort_values('total_SAT', ascending = False).reset_index(drop=True)

top_10_schools = top_10_schools.iloc[0:10]
top_10_schools
```

```python
import numpy as np
borought_largest= schools_no_null.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std']).round(2)

large_std_dev = borought_largest[borought_largest['std'] == borought_largest['std'].max()].reset_index()
largest_std_dev = large_std_dev.rename(columns={'borough':'borough','count':'num_schools','mean':"average_SAT",'std':"std_SAT"})
largest_std_dev
```
